```python
import spacy
import re
import ftfy
def text_standardize(text):
    """
    fixes some issues the spacy tokenizer had on books corpus
    also does some whitespace standardization
    """
    text = text.replace('—', '-')
    text = text.replace('–', '-')
    text = text.replace('―', '-')
    text = text.replace('…', '...')
    text = text.replace('´', "'")
#     text = re.sub(
#         r'''(-+|~+|!+|"+|;+|\?+|\++|,+|\)+|\(+|\\+|\/+|\*+|\[+|\]+|}+|{+|\|+|_+)''', r' \1 ', text)
    text = re.sub(r'\s*\n\s*', ' \n ', text)
    text = re.sub(r'[^\S\n]+', ' ', text)
    return text.strip()
nlp = spacy.blank("zh")
text = "我们可以按照需求来定义进度条的长度，最后一条语句中括号内的数值就可以控制进度条显示的长度，也可以更改时间来定义进度条显示的速度。"
doc = nlp(text_standardize(ftfy.fix_text(text)))
print([t for t in doc])
# print(text_standardize(ftfy.fix_text(text)))
```

```python
def whitespace_tokenize(text):
    """Runs basic whitespace cleaning and splitting on a piece of text."""
    text = text.strip()
    if not text:
        return []
    tokens = text.split()
    return tokens

whitespace_tokenize("晚 风 摇 树 树     还  挺 ")
```

```python
# coding = utf-8
from __future__ import print_function, division
import torch
from torch.utils.data import Dataset, DataLoader
import os
import linecache

class CoupletDataset(Dataset):
    """Couplet Dataset"""
    def __init__(self, x_file, y_file):
        assert os.path.exists(x_file) and os.path.exists(y_file), "the file path is not exists."
        # self.length = len([ "" for line in open(y_file,"r",encoding='utf-8')])
        self.x = linecache.getlines(x_file)
        self.y = linecache.getlines(y_file)
        self.length = len(self.x)

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

couplet_dataset = CoupletDataset(r"data\train\in.txt", r"data\train\out.txt")

dataloader = DataLoader(couplet_dataset, batch_size=512,
                        shuffle=True)

for i_batch, sample_batched in enumerate(dataloader):
    (x, y) = sample_batched
#     print(x)
#     print(y)
#     print(type(x))
    for i in x:
        print(i)

    break
```
